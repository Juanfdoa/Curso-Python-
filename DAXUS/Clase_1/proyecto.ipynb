### Convertir tipo de dato
- int(): Convierte de tipo texto a tipo numerico
- str(): Convierte de tipo numerico a tipo texto

```python
value = 185
print(value, type(value))

text = str(value)
print(text, type(text))

number = int(text)
print(number, type(number))
```

### Proyecto de clase
Se creara un informe presupuestal a partir de la data que ingresara el usuario, se guardara como un arhivo pdf usando una plantilla predefinida
- Instalar fpdf: 'python -m pip install fpdf'


```python
from fpdf import FPDF

proyect = input('Ingrese la descripción del proyecto: ')
estimated_hours = input('Ingrese el total de horas estimadas: ')
hour_value = input('Introduzca el valor de la hora trabajada: ')
delvery_time = input('Introduzca el tiempo estimado de finalización: ')

total_value = int(estimated_hours) * int(hour_value)

pdf = FPDF()
pdf.add_page()
pdf.set_font('Arial')

pdf.image('template.png', x=0, y=0)

pdf.text(115, 145, proyect)
pdf.text(115, 160, estimated_hours)
pdf.text(115, 175, hour_value)

pdf.text(115, 190, delvery_time)
pdf.text(115, 205, str(total_value))

pdf.output('Presupuesto.pdf')
print('¡Presupuesto generado exitosamente!')
```
