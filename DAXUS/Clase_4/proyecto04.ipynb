### Loops / Ciclos
- For
- While

```python
# For
numbers = [1,2,3,4,5]

for number in numbers:
    print(number)
```

```python
# While
counter = 5

while counter > 0:
    print(counter)
    counter = counter-1
```

```python
while True:
    text = input('Escribe algo: ')
    print(text)
    if text == 'parar':
        break
```
