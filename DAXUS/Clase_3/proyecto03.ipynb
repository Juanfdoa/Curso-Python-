### Pandas
- Cargar datos del archivo excel
- Probar algunos metodos

```python
import pandas as pd

# leer archivo excel
data = pd.read_excel("ventas.xlsx")

# Traer los primeros registros
data.head()
```

```python
# Traer los ultimos registros
data.tail()
```

```python
# Cantidad de filas y columnas
data.shape
```

```python
# Informacion sobre las columnas
data.info()
```

```python
# Estadisticas sobre la columna precio
data.describe()
```

```python
# Traer valores unicos de la columna especifica
data['tienda'].unique()
```

```python
# Conteo de valores
data['tienda'].value_counts()
```

```python
# Agrupa por la columna tienda y suma los valores de la columna precio
data.groupby('tienda')['precio'].sum()
```

### Plotly
- Graficos interativos
- Visualizacion de datos

```python
import plotly.express as px

# Ver las tiendas y cuantas veces aparece en la data
px.histogram(data, x='tienda')
```

```python
# Ver tiendas y formas de pago (color)
px.histogram(data,
             x='tienda',
             text_auto=True,
             title='Facturacion',
             color='forma_pago')
```

```python
# Guardar grafico
grafico = px.histogram(data,
             x='tienda',
             text_auto=True,
             title='Facturacion',
             color='forma_pago')

grafico.show()
grafico.write_html('Facturacion.html')
```

```python
# Asegurarte que fecha sea datetime
data['fecha'] = pd.to_datetime(data['fecha'])

# Crear columna año-mes
data['anio_mes'] = data['fecha'].dt.to_period('M').astype(str)

# Agrupando los datos
agrupando = data.groupby(['tienda','anio_mes']).precio.sum().to_frame()

# Reset de los indices
agrupando.reset_index(inplace=True)

# Crear una columna con el valor acumulado
agrupando['acumulado'] = agrupando.groupby('tienda').precio.cumsum()

# Generar grafico
figura = px.bar(agrupando,
                x='acumulado',
                y='tienda',
                color='tienda',
                text_auto=True,
                range_x=[0,160000],
                animation_frame='anio_mes')

figura.show()
```
