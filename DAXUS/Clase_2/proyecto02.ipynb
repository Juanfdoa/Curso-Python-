### YFinance
- Traer datos de cierre del ultimo mes

```python
import yfinance

tiker = input('Ingrese el codigo de la accion: ')

data = yfinance.Ticker(tiker)
table = data.history('1mo')

close = table['Close']
close
```

### Matplotlib
- Generar un gráfico lineal

```python
close.plot()
```

### Estadisticas
- Generar estadisticas (Maxima,Minima,Media)
- Se usa el metodo round para redondear el valor y el 2 para tomar solo 2 decimales

```python
maxima = round(close.max(), 2)
minima = round(close.min(), 2)
media = round(close.mean(), 2)

print(maxima)
print(minima)
print(media)
```

### Proyecto de clase
- Enviar un correo con las estadisticas previamente calculadas

```python
import pyautogui
import pyperclip
import webbrowser
from time import sleep

receiver = 'juanfdoa@hotmail.com'
affair = f'Análisis mensual de la acción {tiker}'
body = f"""
Buen día,

A continuación, se presenta el análisis del comportamiento de la acción '{tiker}' durante el último mes:

Cotización máxima: {maxima} USD
Cotización mínima: {minima} USD
Cotización promedio: {media} USD

Este resumen permite tener una visión general de la evolución del precio en el período analizado.

Quedo atento a cualquier inquietud o información adicional que requieran.
Cordialmente,
"""

# Pausa el sistema cada que pyautogui realice una accion
pyautogui.PAUSE= 2

sleep(1)
# Abre el navegador en la url proporcionanda (Gmail)
webbrowser.open('https://mail.google.com/mail/u/0/?tab=rm&ogbl#inbox')

#Esperar 3 segundos que abra el navehador
sleep(3)

# Hace un click en las cordenadas proporcionandas (x,y)
pyautogui.click(x=89,y=277)

# Completar los campos
pyperclip.copy(receiver)
pyautogui.hotkey("Ctrl","v")
pyautogui.hotkey('Tab')

pyperclip.copy(affair)
pyautogui.hotkey("Ctrl","v")
pyautogui.hotkey('Tab')

pyperclip.copy(body)
pyautogui.hotkey("Ctrl","v")

# Enviar correo
pyautogui.click(x=1232,y=979)

# Cerrar pestaña
pyautogui.hotkey("Ctrl",'F4')

#Resultad0
print('Email enviado exitosamente')
```
