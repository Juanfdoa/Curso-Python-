<h1 style="text-align:center">
 LISTAS POR COMPRENSION
</h1>

Proporcionan una foma concisa de crear y manipular listas usando una sola linea de codigo, son mas legibles y a menudo mas rapidas

### Estructura
- [espression for item in iterable if condition]

```python
squares = [x**2 for x in range(10)]
print(squares)
```

```python
number = [1, 2, 3, 4, 5]
doubled = [x * 2 for x in number]
print(doubled)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7]
evens = [x for x in numbers if x % 2 == 0]
print(evens)
```

```python
names = ['Alice', 'Bob', 'Charlie', 'Dave']
short_names = [name for name in names if len(name) < 5]
print(short_names)
```

```python
numbers = [1, 2, 3, 4, 6]
label = ["Par" if x % 2 == 0 else "Impar" for x in numbers]
print(label)
```

### Ejercicio
- Student Grade Manager

```python
# Step 1: Get studen scores
student_scores = input('Enter student scores separated by commas: ')
scores = [int(score) for score in student_scores.split(",")]

# Step 2: Assign grades using list comprehension
grades = [
    "A" if score >= 90 else
    "B" if score >= 80 else
    "C" if score >= 70 else
    "D" if score >= 60 else
    "F"
    for score in scores
    ]

# Step 3: Filter passing and failing students
passing_students = [score for score in scores if score >= 60]
failing_students = [score for score in scores if score < 60]

# Step 4: Print Result
print('------- Students Grades -------')
for i, (score,grade) in enumerate(zip(scores,grades), start=1):
    print(f"Student {i}: Score= {score}, Grade= {grade}")

print('------- Passing and Failing Students -------')
print('Passing Students:',passing_students)
print('Failing Students:',failing_students)
```
