 MANEJO DE ARCHIVOS
</h1>

Es una forma de almacenar, acceder y manipular datos en archivos en tu computadora

```python
# Leer un archivo, se usa with para asegurar que todo archivo que se abra se pueda cerrar automaticamente
# Como parametros recibe el nombre del archivo y el modo es este caso el modo (r) para lectura
with open("notes.txt","r") as file:
    content = file.read()
    print(content)
```

```python
# Escribir sobre un archivo - SOBREESCRIBE
# Recibe el nombre del archivo sobre el cual escribira y el modo de escritura (w) 
with open("notes.txt", "w") as file:
    file.write("This is a new note.")
```

```python
# Escribir sobre un archivo
# Modo (a) para concatenar en el archivo
with open("notes.txt", "a") as file:
    file.write("\nThis is a second note.")
```

### Ejercicio
- Note-Taking App

```python
# Step 1: Define file name
file_name = 'my_notes.txt'

# Step 2: Display Menu options
def show_menu():
    print('\n------- Note-Taking App Menu -------')
    print('1. Add new note')
    print('2. View notes')
    print('3. Delete notes')
    print('4. Exit')

# Step 3: Add new note
def add_note():
    note = input('Enter your note: ')
    with open(file_name,"a") as file:
        file.write(note + "\n")
    print('Note added successfully!')

# Step 4: View notes
def view_notes():
    try:
        with open(file_name,"r") as file:
            content = file.read()
            if content:
                print("\n------- Your Notes -------")
                print(content)
            else:
                print("\nNo notes found.")
    except FileNotFoundError:
        print("\nNo notes found:")

# Step 5: Delete notes
def delete_notes():
    confirm = input('Are you sure you want to delete all notes? (Yes/No): ')
    if confirm.lower() == "yes":
        with open(file_name,"w") as file:
            pass
        print('All notes have been deleted')
    else:
        print('Deletion Cancelled')


# Step 6: Main Program Loop
while True:
    show_menu()
    
    choice = input('Enter your choice (1-4): ')
    if choice == "1":
        add_note()
    elif choice == "2":
        view_notes()
    elif choice == "3":
        delete_notes()
    elif choice == "4":
        print('Exiting Note-Taking App. Goodbye!')
        break
    else:
        print('Invalid choice. Please enter a number between 1 and 4.')
```
