```python
# Crear diccionario
contact={
    "name":"Jhon Doe",
    "phone":"123-456-789",
    "email":"jhon.doe@email.com"
}

# Imprimir todo el diccionario
print(contact)

# Imprimir una propiedad del diccionario
print(contact["name"])
print(contact.get("name"))
```

```python
# Modificar valor
contact["phone"] = "980-456-8759"
print(contact)
```

```python
# Agregar nuevas entradas
contact["address"] = "123 Main st"
print(contact)
```

```python
# Validar si una clave existe
if "email" in contact:
    # Eliminar un registro
    del contact["email"]
else:
    print("email not found")

print(contact)
```

### Ejercicio
- Contact Book

```python
# Step 1: Inicialize contact List
contacts = {}

# step 2: Create Menu
def show_menu():
    print("------- Contact Book -------")
    print("1. Add contact")
    print("2. View contacts")
    print("3. Search contact")
    print("4. Edit contact")
    print("5. Delete contact")
    print("6. Save contacts on txt file")
    print("7. Exit")

# Step 3: Add contact
def add_contact():
    name = input('Enter contact name: ')
    phone = input('Enter contact phone: ')
    email = input('Enter contact email: ')
    contacts[name] = {"phone":phone, "email":email}
    print(f"contact {name} has been added to your contact book successfully!")

# Step 4: View Contact
def view_contacts():
    if contacts:
        print('----- Contact List -----')
        for name, details in contacts.items():
            print(f"Name: {name}")
            print(f"Phone: {details['phone']}")
            print(f"Email: {details['email']} \n")
    else:
        print('Your contact book is empty.')

# Step 5: Search contact
def search_contact():
    name = input("Enter the contact name: ")
    if name in contacts:
        print(f"Name: {name}")
        print(f"Phone: {contacts[name]['phone']}")
        print(f"Email: {contacts[name]['email']}")
    else:
        print(f"contact {name} not found on your contact book")

# Step 6: Edit contact
def edit_contact():
    name = input("Enter the contact name: ")
    if name in contacts:
        phone = input('Enter new phone number: ')
        email = input('Enter new email')
        contacts[name] = {'phone':phone, 'email':email}
        print(f"Contact {name} has been updated successfully!")
    else:
        print(f"Contact {name} not found in your contact book")

# step 7: Delete contact
def delete_contact():
    name = input('Enter the contact name')
    if name in contacts:
        del contacts[name]
        print(f"Contact {name} has been delete successfully!")
    else:
        print(f"Contact {name} not found in your contact book")

# Step 8: Save contact book
def save_contacts():
    if contacts:
        list_name = 'contacts'
        with open(f"{list_name}.txt", "w") as f:
            for name, details in contacts.items():
                f.write(f"{name}: phone: {details['phone']}, email: {details['email']}\n")
        print(f"The contact book has been saved as {list_name}")
    else:
       print('Your contact book is empty.')

# Step 10: Main program Loop
while True:
    show_menu()
    choice = input('Enter your choice (1-6): ')

    if choice == "1":
        add_contact()
    elif choice == "2":
        view_contacts()
    elif choice == "3":
        search_contact()
    elif choice == "4":
        edit_contact()
    elif choice == "5":
        delete_contact()
    elif choice == "6":
        save_contacts()
    elif choice == "7":
        print('Thank you for using the Contact Book. Goodbye!')
        break
    else:
        print('Invalid choice, Please select a valid option (1-6).')
```
