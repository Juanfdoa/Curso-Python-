### Tuplas
Es una estrutura de datos inmutable, una vez creada no puede ser modificada

```python
# Estructura de tupla
my_tupla = (1,2,3)

# ejemplo
fruits = ('Apple','Banana','Cherry')

# Obtener el primer elemento
print(fruits[0])

# Obtener el ultimo elemento
print(fruits[-1])

# Longitud de la tupla
print(len(fruits))

# Concatenar a la tupla
print(fruits + ('Orange',))
```

```python
coordinates = (10,20,30)
x,y,z = coordinates
print(x)
print(y)
print(z)
```

### Conjuntos
Es una coleccion desordenada de elementos unicos, son mutables pero no pueden ser duplicados

```python
# Estructura de conjunto
my_set = {1,2,3}

# Ejemplo
ingredients = {'Flour', 'Sugar','Butter'}
print(ingredients)

# Agregar dato al conjunto
ingredients.add('Eggs')
print(ingredients)

# Eliminar datos del conjunto
ingredients.remove('Sugar')
print(ingredients)
```

### Operaciones con conjuntos (sets)
- Union: Combina 2 conjuntos diferentes (|)
- Intersection: Elementos comunes en los 2 conjuntos (&)
- Difference: Elementos faltantes en el primer conjunto (-)

```python
set_a = {'Flour', 'Sugar', 'Butter'}
set_b = {'Sugar', 'Eggs'}

# Union
print(set_a | set_b)

# Intersection
print(set_a & set_b)

# Difference
print(set_a - set_b)
```

### Ejercicio
- Ingredients Checker

```python
# Step 1: Define recipe ingredients
recipe_ingredients = {'Flour', 'Sugar', 'Butter', 'Eggs', 'Milk'}

# Step 2: Get user input for available ingredients
user_input = input('Enter the ingredients you have (separated by commas): ')
user_ingredients = set(user_input.split(", "))

# Step 3: Compare ingredients
missing_ingredients = recipe_ingredients - user_ingredients
extra_ingredients = user_ingredients - recipe_ingredients

# Step 4: Display result
print('------- Ingredients Check Result ------')
if missing_ingredients:
    print(f"You are missing the following ingredients: {', '.join(missing_ingredients)}")
else:
    print('You have all the ingredients needed.')

if extra_ingredients:
    print(f"You have extra ingredients: {', '.join(extra_ingredients)}")
else:
    print('You have all the ingredients needed.')
```
