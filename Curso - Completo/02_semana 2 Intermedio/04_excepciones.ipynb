 EXCEPCIONES
</h1>

Son errores que ocurren durante la ejecucion del programa y hacen que el programa se detenga.

### Excepciones comunes en python
- ZeroDivisionError: Error cuando intenta dividir entre 0
- ValueError: Error cuando no se provee el valor correcto 
- Exception: Error inesperado, mas generico

### Palabras reservadas
- Try: Intenta ejecutar el bloque de codigo
- Except: Captura la excepcion y le da manejo
- Else: Se ejecuta si no captura ninguna excepcion
- Finally: Siempre se ejecuta

```python
try:
    num = int(input('Enter a number: '))
    result = 10 / num
    print('Result: ', result)
except ZeroDivisionError:
    print('Error: Division by zero is not allowed')
except ValueError:
    print('Error: invalid input. Please enter a valid number.')
```

```python
try:
    num = int(input('Enter a number: '))
    result = 10 / num
except ZeroDivisionError:
    print('Error: Division by zero is not allowed')
else:
    print('No exception ocurred. Result: ',result)
finally:
    print('Finally block executed. Program Ended')
```

```python
try:
    num = int(input('Enter a number: '))
    result = 10 / num
    print('Result: ', result)
except (ZeroDivisionError, ValueError):
    print('Error: Division by zero is not allowed or invalid input')
```

### Excepciones personalizadas
- Raise

```python
def withdraw(amount):
    if amount < 0:
        raise ValueError('Invalid withdrawal amount - Amount cannot be negative')
    print(f"You have witdrawn ${amount}")

try:
    withdraw(-50)
except ValueError as e:
    print(e)
```

### Ejercicio
- Safe Calculator

```python
# Step 1: Define Calculator functions
def add(x, y):
    return x + y

def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    if y == 0:
        raise ZeroDivisionError('Cannot divide by zero')
    return x / y

# Step 2: Display Menu
def show_menu():
    print('------- Safe Calculator Menu -------')
    print('1. Add')
    print('2. Subtract')
    print('3. Multiplay')
    print('4. Divide')
    print('5. Exit')

# Step 3: Main program
while True:
    show_menu()

    choice = input('Enter your choice (1-5): ')
    if choice == "5":
        print('Exiting the calculator. Goodbye!')
        break

    try:
        num_1 = float(input('Enter first number: '))
        num_2 = float(input('Enter second number: '))

        if choice == "1":
            print('Result:', add(num_1,num_2))
        elif choice == "2":
            print('Result:', subtract(num_1,num_2))
        elif choice == "3":
            print('Result:', multiply(num_1,num_2))
        elif choice == "4":
            print('Result:', divide(num_1,num_2))
        else:
            print('Invalid choice. Please select a valid option (1-5)')
    except ValueError:
        print("Invalid input. Please enter a valid number")
    except ZeroDivisionError as e:
        print(f"Error: {e}")
    except Exception as e:
        print("An unexpected error occurred: {e}")
    finally:
        print("Thank you for using the Safe Calculator!... Restarting...")

    
```
