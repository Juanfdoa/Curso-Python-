Es una coleccion de elementos almacenadas en una sola variable

```python
# Crear lista      0      1       2
shopping_list = ['Milk','Eggs','Bread']

# Imprimir lista
print(shopping_list)

# Imprimir valor segun indice
print(shopping_list[1])
```

### Operaciones y Metodos
- Append('...'): Agregar un valor al final de la lista
- Insert(0,'...'): Agregar un valor a la lista en un indice especifico
- Remove('...'): Elimina un valor en especifico
- Pop(0): Eliminar el ultimo valor, si se le pasa el indice elimina el valor del indice
- Sort(): Ordena la lista en orden alfabetico
- Reverse(): Revierte el orden de la lista
- Clear(): Eliminar todos los elementos de la lista

```python
shopping_list.append('Butter')
print(shopping_list)
```

```python
shopping_list.insert(0,'Orange Juice')
print(shopping_list)
```

```python
shopping_list.remove('Milk')
print(shopping_list)
```

```python
shopping_list.pop()
print(shopping_list)
```

```python
shopping_list.sort()
print(shopping_list)
```

```python
shopping_list.reverse()
print(shopping_list)
```

```python
shopping_list.clear()
print(shopping_list)
```

### Iterar listas
- loop For : Itera el valor de cada item
- loop For (Enumerate): Itera el valor de cada item junto con el indice

```python
for item in shopping_list:
    print(f"- {item}")
```

```python
for index, item in enumerate(shopping_list):
    print(f"{index + 1}. {item}")
```

### Ejercicio
- Shopping List App

```python
import os

# Step 1: Inicialize the list
shop_list = []
 
# Step 2: Define the main Menu
def showMenu():
    print('------- Shopping List Menu -------')
    print('1. View the shopping list')
    print('2. Add an item')
    print('3. Add multiples items from txt file')
    print('4. Edit an item')
    print('5. Remove an item')
    print('6. Save list')
    print('7. Clear list')
    print('8. Exit')

# Step 3: Main Program Loop
while True:
    showMenu()
    choice = input('Enter your choice (1-5): ')

    if choice == "1":
        print('----Shopping List----')
        if not shop_list:
            print('Your shopping list is empty.')
        else:
            for index, item in enumerate(shop_list):
                print(f"{index + 1}. {item}")

    elif choice == "2":
        item = input('Enter the item to add: ')
        shop_list.append(item)
        print(f"{item} has been added to the shopping list.")

    elif choice == "3": 
        list_file = input('Enter the file name: ')
        secondary_list = []
        if os.path.exists(f"{list_file}.txt"):
            with open(f"{list_file}.txt", "r", encoding="utf-8") as f:
                for line in f:
                    shop_list.append(line.strip())
                    secondary_list.append(line.strip())
            print("The items has been added to the shopping list")
            print(secondary_list)
        else:
            print('File was not found, Please try again')
         
    elif choice == "4":
        olditem = input('Enter the item to edit: ')
        if olditem in shop_list:
            newItem = input('Enter the item corrected')
            shop_list[shop_list.index(olditem)] = newItem
            print(f"{olditem} has been updated to {newItem} on the shopping list.")
        else:
            print(f"{olditem} is not on the shopping list, try again")

    elif choice == "5":
        item = input('Enter the item to remove: ')
        if item in shop_list:
            shop_list.remove(item)
            print(f"{item} has been removed from the shopping list")
        else:
            print(f"{item} is not in the shopping list")

    elif choice == "6":
        list_name = 'shopping_list'
        with open(f"{list_name}.txt", "w") as f:
            for item in shop_list:
                f.write(f"{item}\n")
        print(f"The shopping list has been saved as {list_name}")
        
    elif choice == "7":
        shop_list.clear()
        print('The shopping list has been cleared')

    elif choice == "8":
        print('Goodbye! Happy Shopping')
        break

    else:
        print('Invalid choice. Please try again')
```
